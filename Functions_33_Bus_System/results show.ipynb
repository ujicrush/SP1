```python
import numpy as np
import matplotlib.pyplot as plt
```

### single

```python
res_path = 'res/single'

obj_2nd_s = np.load(f"{res_path}/obj_2nd.npy")
lambda_2nd_s = np.load(f"{res_path}/lambda_2nd.npy")
mu_2nd_s = np.load(f"{res_path}/mu_2nd.npy")
previous_rating_s = np.load(f"{res_path}/previous_rating.npy")
previous_cap_s = np.load(f"{res_path}/previous_cap.npy")
ESS_loc_s = np.load(f"{res_path}/ESS_loc.npy")
alpha_store_s = np.load(f"{res_path}/alpha_store.npy")
upperB_save_s = np.load(f"{res_path}/upperB_save.npy")
lowerB_save_s = np.load(f"{res_path}/lowerB_save.npy")
invest_save_s = np.load(f"{res_path}/invest_save.npy")
total_time_s = np.load(f"{res_path}/total_time.npy", allow_pickle=True)

print("obj_2nd shape:", obj_2nd_s.shape)
print("lambda_2nd shape:", lambda_2nd_s.shape)
print("mu_2nd shape:", mu_2nd_s.shape)
print("previous_rating shape:", previous_rating_s.shape)
print("previous_cap shape:", previous_cap_s.shape)
print("ESS_loc shape:", ESS_loc_s.shape)
print("alpha_store shape:", alpha_store_s.shape)
print("upperB_save shape:", upperB_save_s.shape)
print("lowerB_save shape:", lowerB_save_s.shape)
print("total_time shape:", total_time_s.shape)
```

```python
iterations_s = np.arange(len(upperB_save_s))

plt.figure(figsize=(5, 3))
plt.plot(iterations_s, upperB_save_s, marker='o', label='Upper Bound', linewidth=2)
plt.plot(iterations_s, lowerB_save_s, marker='s', label='Lower Bound', linewidth=2)
plt.title("Convergence of Upper and Lower Bounds", fontsize=12)
plt.xlabel("Iteration", fontsize=10)
plt.ylabel("Convergence Process", fontsize=10)
plt.xticks(iterations_s) 
plt.legend(fontsize=8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
